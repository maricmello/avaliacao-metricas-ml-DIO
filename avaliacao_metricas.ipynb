Importando bibliotecas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Funções para o cálculo das métricas

```python
#Sensibilidade
def calcular_sensibilidade(VP,FN):
    sensibilidade = VP/(VP+FN)
    return sensibilidade

#Especificidade
def calcular_especificidade(VN,FP):
    especificidade = VN/(FP+VN)
    return especificidade

#Acurácia
def calcular_acuracia(VP,VN,FN,FP):
    acuracia = (VP+VN)/(VP+FN+FP+VN)
    return acuracia

#Precisão
def calcular_precisao(VP,FP):
    precisao = VP/(VP+FP)
    return precisao

#F score
def calcular_fscore(precisao,sensibilidade):
    fscore = 2 * (precisao * sensibilidade)/(precisao*sensibilidade)
    return fscore
```

Exemplo

```python
# Valores arbitrários para matriz de confusão (exemplo)
VP = 70
VN = 100
FP = 10
FN = 20

# Cálculo das métricas
acuracia = calcular_acuracia(VP, FN,FP,VN)
sensibilidade = calcular_sensibilidade(VP, FN)
especificidade = calcular_especificidade(VN, FP)
precisao = calcular_precisao(VP, FP)
fscore = calcular_fscore(precisao, sensibilidade)


print(f"Acurácia: {acuracia:.3f}")
print(f"Sensibilidade (Recall): {sensibilidade:.3f}")
print(f"Especificidade: {especificidade:.3f}")
print(f"Precisão: {precisao:.3f}")
print(f"F-Score: {fscore:.3f}")
```

Matriz de confusão (Gráfico)

```python
matriz = np.array([[VP, FN],
                   [FP, VN]])

df_cm = pd.DataFrame(
    matriz, 
    index=["Verdadeiro Positivo", "Verdadeiro Negativo"], 
    columns=["Negativo Positivo", "Positivo Negativo"]
)

plt.figure(figsize=(6,4))
sns.heatmap(df_cm, annot=True, fmt='d', cmap="Blues", cbar=False)
plt.title("Matriz de Confusão")
plt.ylabel("Classe Real")
plt.xlabel("Classe Prevista")
plt.show()
```
